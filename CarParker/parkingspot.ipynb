```python
import numpy as np
import pandas as pd
import os
import PIL
from PIL import Image
import matplotlib.pyplot as plt
from matplotlib.image import imread
import cv2
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
```

```python
# Define the directory containing the dataset
data_directory = "/kaggle/input/find-a-car-park/data/"

# Define the categories for parking states
parking_states = ["Free", "Full"]

# Set the desired image size
image_dimension = 150
```

```python
import os
import cv2

car_park_dataset = []

def construct_dataset():
    for state in parking_states:
        state_path = os.path.join(data_directory, state)
        state_label = parking_states.index(state)
        
        print("Processing state:", state)
        for image_file in os.listdir(state_path):
            print("  Reading image:", image_file)
            image_path = os.path.join(state_path, image_file)
            
            # Read and resize the image
            image_array = cv2.imread(image_path)
            resized_image_array = cv2.resize(image_array, (image_dimension, image_dimension))
            
            # Append the resized image and its label to the dataset
            car_park_dataset.append([resized_image_array, state_label])

# Construct the dataset
construct_dataset()
```

```python
# Initialize lists to store features (X) and labels (Y)
parking_features = []
parking_labels = []

# Extract features and labels from the dataset
for features, label in car_park_dataset:
    parking_features.append(features)
    parking_labels.append(label)
```

```python
# Split the dataset into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(parking_features, parking_labels, test_size=0.20, random_state=42)
```

```python
import tensorflow as tf

def preprocess_and_resize(input_image):
    # Normalize the input image
    normalized_input_image = tf.image.per_image_standardization(input_image)
    # Resize the image to the desired dimensions
    resized_image = tf.image.resize(normalized_input_image, (227, 227))
    return resized_image
```

```python
X_train_processed = preprocess_and_resize(X_train)
X_test_processed = preprocess_and_resize(X_test)
```

```python
X_train_array = np.array(X_train_processed).reshape(-1, 227, 227, 3)
Y_train_array = np.array(Y_train)
X_test_array = np.array(X_test_processed).reshape(-1, 227, 227, 3)
Y_test_array = np.array(Y_test)
```

```python
X_validation = X_train_array[:520]
X_train_array = X_train_array[521:]
Y_validation = Y_train_array[:520]
Y_train_array = Y_train_array[521:]
```

```python
from tensorflow import keras

# Define a Sequential model
parking_spot_model = keras.models.Sequential([
    # First convolutional layer
    keras.layers.Conv2D(filters=96, kernel_size=(11, 11), strides=(4, 4), activation='relu', input_shape=(227, 227, 3)),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(3, 3), strides=(2, 2)),
    
    # Second convolutional layer
    keras.layers.Conv2D(filters=256, kernel_size=(5, 5), strides=(1, 1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(3, 3), strides=(2, 2)),
    
    # Third and fourth convolutional layers
    keras.layers.Conv2D(filters=384, kernel_size=(3, 3), strides=(1, 1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.Conv2D(filters=384, kernel_size=(3, 3), strides=(1, 1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    
    # Fifth convolutional layer
    keras.layers.Conv2D(filters=256, kernel_size=(3, 3), strides=(1, 1), activation='relu', padding="same"),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(3, 3), strides=(2, 2)),
    
    # Flatten layer
    keras.layers.Flatten(),
    
    # First fully connected layer
    keras.layers.Dense(4096, activation='relu'),
    keras.layers.Dropout(0.5),
    
    # Second fully connected layer
    keras.layers.Dense(4096, activation='relu'),
    keras.layers.Dropout(0.5),
    
    # Output layer
    keras.layers.Dense(10, activation='softmax')
])

# Display the model summary
parking_spot_model.summary()
```

```python
# Compile the model with specified loss, optimizer, and metrics
parking_spot_model.compile(loss='sparse_categorical_crossentropy', 
                           optimizer=tf.optimizers.Adam(learning_rate=0.001), 
                           metrics=['accuracy'])

# Display the model summary
parking_spot_model.summary()
```

```python
from tensorflow.keras import callbacks

# Define early stopping callback
parking_early_stopping = callbacks.EarlyStopping(
    monitor="val_loss",
    mode="min",
    patience=5,
    restore_best_weights=True
)
```

```python
# Fit the model
parking_history = parking_spot_model.fit(
    X_train_array,
    Y_train_array,
    epochs=25,
    validation_data=(X_validation, Y_validation),
    callbacks=[parking_early_stopping]
)
```

```python
# Evaluate the model on the test data
test_loss, test_accuracy = parking_spot_model.evaluate(X_test_array, Y_test_array)
```
